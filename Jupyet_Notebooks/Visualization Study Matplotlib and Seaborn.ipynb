# Matplotlib Practice

```python
import math
x=np.arange(0,math.pi*2,0.05)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel("Angle")
plt.ylabel("Sine Value")
plt.title("The Sinusoidal Wave")
plt.show()
```

## To Tweak the size and dpi you can use the fig and ax!! :)

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
fig = plt.figure(figsize =(5, 4))
ax = fig.add_axes([0,0,1,1])
#left, bottom, width and height 
ax.plot(x,y)
fig.set_dpi(150.0)
ax.set_title("sine wave")
ax.set_xlabel('angle')
ax.set_ylabel('sine')
plt.show()
```

```python
y = [1, 4, 9, 16, 25,36,49, 64]
x1 = [1, 16, 30, 42,55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
l1=ax.plot(x1,y)
l2=ax.plot(x2,y)
ax.legend(labels=('tv','smartphone'),loc='upper left')
plt.xlabel('points')
plt.ylabel('points again')
plt.title('the normal graph with interative functions')
plt.show()
```

## Creating mutliple plots on same axes

```python
import matplotlib.pyplot as plt
import numpy as np
import math
x = np.arange(0, math.pi*2, 0.05)
fig=plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.55, 0.55, 0.3, 0.3]) # inset axes
y = np.sin(x)
axes1.plot(x, y, 'b')
axes2.plot(x,np.cos(x),'r')
axes1.set_title('sine')
axes2.set_title("cosine")
plt.show()
```

## In the following example, the plot has dual y axes, one showing exp(x) and the other showing log(x) 

```python
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
x = np.arange(1,11)
a1.plot(x,np.exp(x))
a1.set_ylabel('exp')
a2 = a1.twinx()
a2.plot(x, np.log(x),'ro-')
a2.set_ylabel('log')
fig.legend(labels = ('exp','log'),loc='upper left')
plt.show()
```
